Importing Dataset

```python
import csv
with open('insurance.csv', newline='') as insurance_csv:
    full_contents = insurance_csv.read()
    print(full_contents)
```

Extracting the features of the dataset

```python
#Empty lists to store the features
ages = []
sex = []
bmi = []
children = []
smoker = []
region = []
charges = []

#using DictReader() to put these values into the lists
with open('insurance.csv', newline='') as patient_csv:
    patient_info = csv.DictReader(patient_csv)
    for row in patient_info:
        ages.append(row['age'])
        sex.append(row['sex'])
        bmi.append(row['bmi'])
        children.append(row['children'])
        smoker.append(row['smoker'])
        region.append(row['region'])
        charges.append(row['charges'])

print(ages)
print(sex)
print(bmi)
print(children)
print(smoker)
print(region)
print(charges)
```

DATA ANALYSIS

Finding average age

```python
#finding average age
def f_average_age(age_list):
    age_counter = 0
    for age in age_list:
        age_counter += int(age)
    return int(age_counter / len(age_list))
f_average_age(ages)
```

Finding Male vs. Female ratio

```python
def sex_ratio(gender_list):
    male_counter = 0
    female_counter = 0
    for gender in gender_list:
        if gender == 'male':
            male_counter += 1
        else:
            female_counter += 1
    ratio = male_counter / female_counter
    if ratio > 1:
        print("Male individuals are " + str((ratio - 1)*100) + " percent higher than female in the sample.")
    else:
        print("Female individuals are " + str((1 - ratio)*100) + " percent higher than female in the sample")
    print("Male gender represent " + str((male_counter / len(gender_list)) * 100) + " percent of the sample")
    print("Female gender represent " + str((female_counter / len(gender_list)) * 100) + " percent of the sample")

sex_ratio(sex)
```

Defining what's the influence of having children

```python
child_cost = list(zip(children, charges))
print(child_cost)
```

```python
def children_influence(num_children):
    zero_children = []
    one_children = []
    two_children = []
    three_children = []
    plus_four_children = []
    for parent in num_children:
        if parent[0] == '0':
            zero_children.append(float(parent[1]))
        elif parent[0] == '1':
            one_children.append(float(parent[1]))
        elif parent[0] == '2':
            two_children.append(float(parent[1]))
        elif parent[0] == '3':
            three_children.append(float(parent[1]))
        else:
            plus_four_children.append(float(parent[1]))
    print("Insurance costs for patients with no children goes from " + str(min(zero_children)) + " to " + str(max(zero_children)))
    print("Insurance costs for patients with 1 child goes from " + str(min(one_children)) + " to " + str(max(one_children)))
    print("Insurance costs for patients with 2 children goes from " + str(min(two_children)) + " to " + str(max(two_children)))
    print("Insurance costs for patients with 3 children goes from " + str(min(three_children)) + " to " + str(max(three_children)))
    print("Insurance costs for patients with 4 or more children goes from " + str(min(plus_four_children)) + " to " + str(max(plus_four_children)))
    print("Patients with no kids: " + str(len(zero_children)))
    print("Patients with one kid: " + str(len(one_children)))
    print("Patients with two kids: " + str(len(two_children)))
    print("Patients with three kids: " + str(len(three_children)))
    print("Patients with four or more kids: " + str(len(plus_four_children)))
    one_at_least = (len(one_children) + len(two_children) + len(three_children) + len(plus_four_children))
    total_children = (len(zero_children) + len(one_children) + len(two_children) + len(three_children) + len(plus_four_children))
    print("Patients with at least one kid: " + str(one_at_least) + " (" + str((one_at_least / total_children)*100) + " percent of the sample.)")
children_influence(child_cost)
```

Average age for men and women

```python
def avg_age_gender(age, sex):
    age_sex_zip = list(zip(age, sex))
    male_counter = 0
    female_counter = 0
    female_age_counter = 0
    male_age_counter = 0
    for pair in age_sex_zip:
        if pair[1] == 'male':
            male_age_counter += int(pair[0])
            male_counter += 1
        else:
            female_age_counter += int(pair[0])
            female_counter += 1
    avg_age_female = female_age_counter / female_counter
    avg_age_male = male_age_counter / male_counter
    print("The average age for female patients is: " + str(round(avg_age_female,2)))
    print("The average age for male patients is: " + str(round(avg_age_male,2)))

avg_age_gender(ages, sex)
```

Average charge for men and women

```python
def avg_charge_sex(sex, charges):
    sex_charges_zip = list(zip(sex, charges))
    male_counter = 0
    female_counter = 0
    charge_female_counter = 0
    charge_male_counter = 0
    for pair in sex_charges_zip:
        if pair[0] == 'male':
            charge_male_counter += float(pair[1])
            male_counter += 1
        else:
            charge_female_counter += float(pair[1])
            female_counter += 1
    avg_charge_female = charge_female_counter / female_counter
    avg_charge_male = charge_male_counter / male_counter
    print("The average cost of insurance for female patients is: " + str(avg_charge_female))
    print("The average cost of insurance for male patients is: " + str(avg_charge_male))

avg_charge_sex(sex,charges)
```

What's the average cost according to BMI?

```python
def bmi_category_charge(bmi, charges):
    bmi_charges_zip = list(zip(bmi, charges))
    bmi_charges_category = []
    #assigning a category on each patient according to their BMI
    for patient in bmi_charges_zip:
        if float(patient[0]) < 18.5:
            bmi_charges_category.append([patient[0], patient[1], 'underweight'])
        elif 18.5 <= float(patient[0]) <= 24.9:
            bmi_charges_category.append([patient[0], patient[1], 'Normal weight'])
        elif 25.0 <= float(patient[0]) <= 29.9:
            bmi_charges_category.append([patient[0], patient[1], 'Overweight'])
        elif 30.0 <= float(patient[0]) <= 34.9:
            bmi_charges_category.append([patient[0], patient[1], 'Obesity Class I (Moderate)'])
        elif 35.0 <= float(patient[0]) <= 39.9:
            bmi_charges_category.append([patient[0], patient[1], 'Obesity Class II (Severe)'])
        else:  # bmi >= 40.0
            bmi_charges_category.append([patient[0], patient[1], 'Obesity Class III (Very severe or morbid obesity)'])
    #counting patients on each category
    uw_counter = 0
    nw_counter = 0
    ow_counter = 0
    oc1_counter = 0
    oc2_counter = 0
    oc3_counter = 0
    for patient in bmi_charges_category:
        if patient[2] == 'underweight':
            uw_counter += 1
        elif patient[2] == 'Normal weight':
            nw_counter += 1
        elif patient[2] == 'Overweight':
            ow_counter += 1
        elif patient[2] == 'Obesity Class I (Moderate)':
            oc1_counter += 1
        elif patient[2] == 'Obesity Class II (Severe)':
            oc2_counter += 1
        else: #patient[2] == 'Obesity Class III (Very severe or morbid obesity)':
            oc3_counter += 1
    print("Number of underweight patients: " + str(uw_counter))
    print("Number of patients with normal weight: " + str(nw_counter))
    print("Number of overweighed patients: " + str(ow_counter))
    print("Number of patients with Obesity Class I: " + str(oc1_counter))
    print("Number of patients with Obesity Class II: " + str(oc2_counter))
    print("Number of patients with Obesity Class III: " + str(oc3_counter))
    uw_cost = 0
    nw_cost = 0
    ow_cost = 0
    oc1_cost = 0
    oc2_cost = 0
    oc3_cost = 0
    for patient in bmi_charges_category:
        if patient[2] == 'underweight':
            uw_cost += float(patient[1])
        elif patient[2] == 'Normal weight':
            nw_cost += float(patient[1])
        elif patient[2] == 'Overweight':
            ow_cost += float(patient[1])
        elif patient[2] == 'Obesity Class I (Moderate)':
            oc1_cost += float(patient[1])
        elif patient[2] == 'Obesity Class II (Severe)':
            oc2_cost += float(patient[1])
        else: #patient[2] == 'Obesity Class III (Very severe or morbid obesity)':
            oc3_cost += float(patient[1])
    uw_cost_avg = round((uw_cost / uw_counter),3)
    nw_cost_avg = round((nw_cost / nw_counter),3)
    ow_cost_avg = round((ow_cost / ow_counter),3)
    oc1_cost_avg = round((oc1_cost / oc1_counter),3)
    oc2_cost_avg = round((oc2_cost / oc2_counter),3)
    oc3_cost_avg = round((oc3_cost / oc3_counter),3)
    print("The average cost for underweight patients is: " + str(uw_cost_avg))
    print("The average cost for Normal weight patients is: " + str(nw_cost_avg))
    print("The average cost for overweight patients is: " + str(ow_cost_avg))
    print("The average cost for obesity1 patients is: " + str(oc1_cost_avg))
    print("The average cost for obesity2 patients is: " + str(oc2_cost_avg))
    print("The average cost for obesity3 patients is: " + str(oc3_cost_avg))
    #organizing into dictionaries
    patient_bmi_dict = {
                            "Underweight": uw_counter,
                            "Normal weight": nw_counter,
                            "Overweight": ow_counter,
                            "Obesity Class I": oc1_counter,
                            "Obesity Class II": oc2_counter,
                            "Obesity Class III": oc3_counter
    }
    print(patient_bmi_dict)
    avgcost_bmi_dict = {}
    for category in patient_bmi_dict:
        if category == "Underweight":
            avgcost_bmi_dict[category] = uw_cost_avg
        elif category == "Normal weight":
            avgcost_bmi_dict[category] = nw_cost_avg
        elif category == "Overweight":
            avgcost_bmi_dict[category] = ow_cost_avg
        elif category == "Obesity Class I":
            avgcost_bmi_dict[category] = oc1_cost_avg
        elif category == "Obesity Class II":
            avgcost_bmi_dict[category] = oc2_cost_avg
        else: #category == "Obesity Class III":
            avgcost_bmi_dict[category] = oc3_cost_avg
    print(avgcost_bmi_dict)
bmi_category_charge(bmi,charges)

```

Influence of smoking on insurance cost

```python
def smoking_cost(smoker, charges):
    smoker_charges_zip = list(zip(smoker, charges))
    smoker_counter = 0
    non_smoker_counter = 0
    smoker_charge_counter = 0
    non_smoker_charge_counter = 0
    for patient in smoker_charges_zip:
        if patient[0] == 'yes':
            smoker_charge_counter += float(patient[1])
            smoker_counter += 1
        else: #patient[0] == 'no'
            non_smoker_charge_counter += float(patient[1])
            non_smoker_counter += 1
    print("The average cost of insurance for patients who smoke is: " + str(smoker_charge_counter / smoker_counter))
    print("The average cost of insurance for patients who smoke is: " + str(non_smoker_charge_counter / non_smoker_counter))

smoking_cost(smoker, charges)
```
